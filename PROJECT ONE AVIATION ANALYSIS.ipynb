CLEANING THE DTASET

Some columns in the dataset aren’t helpful for my analysis, mostly because they have too many missing values or don’t relate to aircraft risk. I’ll start by dropping those.

For injury columns like "Total.Fatal.Injuries", "Total.Minor.Injuries", "Total.Serious.Injuries", and "Total.Uninjured", I’m going to fill any missing values with 0. This is because if the injury field was left blank, it probably means no injuries were reported.

I’m also keeping important columns like "Aircraft.damage", "Aircraft.Category", and "Weather.Condition", even though they have some missing data .

```python
#drop the column that have too many missing values that aren't relevant
df.drop(columns=['Schedule','Air.carrier','FAR.Description','Longitude','Latitude','Airport.Code','Airport.Name','Publication.Date'], inplace=True)
```

```python
#fill in the columns with missing injury values with zero
df['Total.Fatal.Injuries'] = df['Total.Fatal.Injuries'].fillna(0)
df['Total.Minor.Injuries'] = df['Total.Minor.Injuries'].fillna(0)
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(0)
df['Total.Uninjured'] = df['Total.Uninjured'].fillna(0)
```

REMOVE THE WHITESPACES
Some columns in the dataset may contain extra spaces either at the beginning or end of the text.

To fix this, I’ll remove white spaces from:
- All column names
- All string (object) values in the dataset

```python
# Remove leading spaces from column names
df.columns = df.columns.str.strip()

# Remove spaces from text entries in all object (string) columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

COMPARE AIRCRAFT DAMAGE BY CATEGORY
I will look at how damage levels differ between aircraft categories. This can help show which aircraft tend to have more serious accidents.

The column "Aircraft.damage" contains values like "Destroyed", "Substantial", and "Minor". I’ll group the data by both "Aircraft.Category" and "Aircraft.damage" to see how often each type of damage happened for each aircraft type.

```python
# Group by aircraft type and damage level
damage_counts = df.groupby(['Aircraft.Category', 'Aircraft.damage']).size().unstack().fillna(0)

# Show the result
damage_counts
```

ABOUT THE DAMAGE RESULS ANALYSIS
This shows us how much damage is involved when different aircraft types are in accidents.

If one type has a high number of "Destroyed" outcomes compared to others, that might be more risky.

VISUALIZATION FOR AVERAGE FATAL INJURIES

```python
import seaborn as sns
#Group for seaborn
avg_fatal = df.groupby('Aircraft.Category')['Total.Fatal.Injuries'].mean().reset_index()

# Sort the values
avg_fatal = avg_fatal.sort_values('Total.Fatal.Injuries', ascending=False)

# Create seaborn barplot
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_fatal, x='Aircraft.Category', y='Total.Fatal.Injuries', palette='Blues')

plt.title('Average Fatal Injuries by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Average Fatal Injuries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

VISUALIZATION OF AIRCRAFT DAMAGE BY CATEGORY 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group and count data
damage_data = df.groupby(['Aircraft.Category', 'Aircraft.damage']).size().reset_index(name='Count')

# create the grouped bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=damage_data, x='Aircraft.Category', y='Count', hue='Aircraft.damage', palette='Set2')

# Customize with Matplotlib
plt.title('Aircraft Damage by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Damage Level')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
