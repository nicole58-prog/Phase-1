DATA SOURCE
The data is stored in the file 'AviationData.csv'. It includes historical records of aviation incidents and requires special encoding to be read correctly.

```python
#import pandas
import pandas as pd
```

```python
#read the aviation data csv file with proper encoding
df = pd.read_csv('AviationData.csv', encoding='latin1')
```

DATA CLEANING

```python
#display the first five rows of the dataset
df.head()
```

```python
#read the summary information of the dataset
df.info()
```

```python
#describe the dataset to get statistical information
df.describe()
```

```python
#drop any duplicate rows if any
df.drop_duplicates()
```

```python
#count the number of missing values in each column
df.isnull().sum()
```

```python
#print actual column names
print(df.columns.tolist())
```

```python
#strip whitespaces from column names
df.columns.str.strip()
```

```python
#drop columns with too many missing values, ignore if not present
df = df.drop(columns=['Latitude', 'Longitude', 'Airport.Code', 'Airport.Name', 'FAR.Description', 'Schedule', 'Air.carrier','Publication.Date'], errors='ignore')
df.columns
```

DATA IMPUTATION

```python
#fill the remaining injury related columns 
df['Total.Minor.Injuries'] = df['Total.Minor.Injuries'].fillna(0)
df['Total.Uninjured'] = df['Total.Uninjured'].fillna(0)
df['Registration.Number'] = df['Registration.Number'].fillna('Unknown')
```

```python
#confirm if there are no missing values
df.isnull().sum()
```

DATA ANALYSIS

```python
#read rows and columns
df.shape
```

```python
# Top categories for Aircraft Category
df['Aircraft.Category'].value_counts()
```

```python
#grouped analysis
# Average fatal injuries per aircraft category
df.groupby('Aircraft.Category')['Total.Fatal.Injuries'].mean().sort_values(ascending=False)

# Total fatal injuries by flight phase
df.groupby('Broad.phase.of.flight')['Total.Fatal.Injuries'].sum().sort_values(ascending=False)
```

DATA VISUALIZATION

```python
import matplotlib.pyplot as plt

# Create a Total.Injuries column if it doesn't exist
if 'Total.Injuries' not in df.columns:
	df['Total.Injuries'] = (df['Total.Fatal.Injuries'].fillna(0) + df['Total.Serious.Injuries'].fillna(0) + df['Total.Minor.Injuries'].fillna(0))
	
# group by category and sum of total injuries
injuries_by_category = df.groupby('Aircraft.Damage')['Total.Injuries'].sum()


# plot the graph
plt.figure(figsize=(10,6))
plt.bar(injuries_by_category.index, injuries_by_category.values, color='blue')
plt.title('Total Injuries by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Total Injuries')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()


```
